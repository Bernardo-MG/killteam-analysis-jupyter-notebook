# Kill Team Damage Calculations

```python
import pandas as pd
from scripts.dice import average_dice_value
from scripts.values import user_value
from scripts.killteam import chance_to_hit, chance_to_wound, chance_to_fail_armour, chance_to_damage, chance_to_damage_at_least_once

weapons_average = pd.read_csv('datasets/1st/weapons.csv')
weapons_average = weapons_average.sort_values(by=['name'])
```

```python
average_strength = 3
average_attacks = 1
average_ballistics = 3
average_toughness = 3
average_armor = 1
```

## Averages

```python
weapons_average["attacks"] = weapons_average.apply(lambda x: average_dice_value(x["attacks"]), axis=1)
weapons_average["damage"] = weapons_average.apply(lambda x: average_dice_value(x["damage"]), axis=1)
```

```python
weapons_average["attacks"] = weapons_average.apply(lambda x: user_value(x["attacks"], average_attacks), axis=1)
weapons_average["strength"] = weapons_average.apply(lambda x: user_value(x["strength"], average_strength), axis=1)
```

```python
weapons_average
```

## Hit chance

```python
weapons_average["to_hit"] = weapons_average.apply(lambda x: chance_to_hit(average_ballistics), axis=1)
weapons_average["to_wound"] = weapons_average.apply(lambda x: chance_to_wound(int(x["strength"]), average_toughness), axis=1)
weapons_average["to_fail_armor"] = weapons_average.apply(lambda x: chance_to_fail_armour(int(x["ap"]), average_armor), axis=1)
weapons_average["to_damage"] = weapons_average.apply(lambda x: chance_to_damage(float(x["to_hit"]), float(x["to_wound"]), float(x["to_fail_armor"])), axis=1)
weapons_average["to_damage_at_least_once"] = weapons_average.apply(lambda x: chance_to_damage_at_least_once(float(x["to_damage"]), float(x["attacks"])), axis=1)
```
